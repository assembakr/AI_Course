### problem 2 : Compare between Adjacency Matrix/List

###### Adjacency Matrix:

It's way to represent a **graph** in computer.
Also, it's a matrix A of n*n;  where n is the number of vertices <br>
a [j] [k] = 1; if j & k are adiacent, or 
        = 0; if not adjacent <br>
**complexity:** As the Array should be **n*n**, so in worst case the space required will be O(|n|^2) <br>
the adjacency matric is **better** for **dense graph** (almost all the nodes are connected to each other)

###### Adjacency List:
it's a collection of unorderd lists used to represt a graph in computer <br>

**complexity:** for every vertex we store its neighbours. In the worst case, if a graph is connected O(n) is required for a vertex and O(E) is required for storing neighbours corresponding to every vertex .Thus, overall space complexity is O(|n|+|E|).<br>
the adjacency list is **better** for **spare graph**

### problem 3 : BFS and DFS algorithms

<h3> BFS: </h3> 

**Definition:** stands for breadth first search, it's a traversal approach in which we first walk theought all nodes on the same level before moving on the next node. <br>
**Data Structure:** it uses the queue to find the shorted path. <br>
**Speed:** slower than the DFS. <br>
**Time Complexity:** Time Complexity of BFS depends on it's a adjacency matrix or adjacency list. for adjacency list the complexity will be = O(V+E) where V is vertices and E is edges, or will be = V<sup>2</sup> if it was adjacency matrix <br>
**Usage:** when the target is close. <br>
**suitability for decision tree:** not suitable for decision trees because it considers all neighbours.

<h3> DFS: </h3>

**Definition:** stands for Deapth First Search, it's a traversal approach in which the traverse begins at the root node and proceeds throught the nodes as far as pssible until we reach a node with no unvisited nearby nodes. <br>
**Data Structure:** it uses the stack. <br>
**Speed:** faster than the DFS. <br>
**Time Complexity:** Time Complexity of BFS depends on it's a adjacency matrix or adjacency list. for adjacency list the complexity will be = O(V+E) where V is vertices and E is edges, or will be = V<sup>2</sup> if it was adjacency matrix <br>
**Usage:** when the target is far. <br>
**suitability for decision tree:** it's more suitable for decision tree. As with one decision, we need to traverse further to augment the decision. If we reach the conclusion, we won..

### problem 5: create a function recives stack elements as inputs from the user while not equal zero

```python
from queue import LifoQueue
stack = LifoQueue()
def recives():
    
    x = float(input("Enter a number: "))
    
    while (x != 0):
        stack.put(x)
        x = float(input("Enter a number: ")) #to keep reciving inputs until he put zero.
    
    #the stack revives no inputs any more.
    if x == 0:
        print("The stack is done!")
        
    return stack

#call the function        
recives()
```
